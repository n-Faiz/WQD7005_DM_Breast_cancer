```python
X = df.iloc[:,2:]
y = df.iloc[:,1]
X.to_csv('test.csv')
```

# Exploratory Data Analysis / Visualization

```python
# Exploratory - unsupervised 02

fig, axes = plt.subplots(15, 2, figsize=(20, 40))
malignant = X[y == 1]
benign = X[y == 0]
ax = axes.ravel()
for i in range(30):
    _, bins = np.histogram(X.iloc[:, i], bins=50)
    ax[i].hist(malignant.iloc[:, i], bins=bins, color='red', alpha=.5)
    ax[i].hist(benign.iloc[:, i], bins=bins, color='green', alpha=.5)
    ax[i].set_title(X.columns[i])
    ax[i].set_yticks(())

ax[0].set_xlabel("Feature magnitude")
ax[0].set_ylabel("Frequency")
ax[0].legend(["malignant", "benign"], loc="best")
fig.tight_layout()
plt.show()
```

# Modeling

```python
# train-split test

from sklearn.model_selection import train_test_split
Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, random_state=42 ,train_size = 0.7)
```

```python
# Feature scaling

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(Xtrain)

Xtrain_s = scaler.transform(Xtrain)
Xtest_s = scaler.transform(Xtest)
```

```python
# modeling - unsupervised 02
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier

from sklearn import metrics
from sklearn.metrics import confusion_matrix


svc = SVC(kernel='rbf')
logistic_reg = LogisticRegression()
knn = KNeighborsClassifier(n_neighbors=5)
nb = GaussianNB()
d_tree = DecisionTreeClassifier(max_depth=30)
mlp = MLPClassifier(hidden_layer_sizes=(5, ),max_iter = 500, random_state=42)

models = ['svc','logistic_reg', 'knn', 'nb', 'd_tree','mlp']
```

```python
# modeling
model_name = ['Support Vector Machines',
              'Logistic Regression',
              'K neighbour Classification',
              'Naive bayes',
              'Decision Tree','Multi-layer Perceptron']
score = pd.DataFrame(columns = ["Model","Accuracy Score" ,"Recall Score","Precision Score:","f1 score"] )

for i in range(len(models)):
    model = eval(models[i]).fit(Xtrain_s,ytrain)
    pred = eval(models[i]).predict(Xtest_s)

    score = score.append({
        "Model":model_name[i],
        "Accuracy Score" : metrics.accuracy_score(ytest,pred),
        "Recall Score" : metrics.recall_score(ytest,pred),
        "Precision Score:": metrics.precision_score(ytest,pred),
        "f1 score":metrics.f1_score(ytest,pred)
    },ignore_index = True)
```

# Scientific findings

Here we can see that in term of <u>recall score</u>, **MLP** scores the highest, followed by logistic regression. For the evaluation score, the score for recall is prioritize, as it is important that we do not classify the malignant tumor to be benign tumor. This is because it is impertinent that the malignant tumor patients get the appropriate treatment, while it is optional only for benign tumor patients to receive it.
